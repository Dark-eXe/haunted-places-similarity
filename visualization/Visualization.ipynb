# Cluster Visualizations

```python
import pandas as pd
```

```python
import os
import shutil
```

#### Parameters

```python
metric = "edit" # jaccard | cos | edit
partition = 199
```

### Move Cluster Files to Current Directory

```python
# Get the current working directory
current_directory = os.getcwd()

cluster_files = ["circle.json", "clusters.json", "levelCluster.json"]
for file in cluster_files:
    # Source file path
    source_file = os.path.join("..", "results", "clusters", f"dir_{partition}", metric, file)

    # Destination file path (current directory)
    destination_file = os.path.join(current_directory, os.path.basename(source_file))
    
    # Copy the file
    shutil.copy(source_file, destination_file)
```
